```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path

# Files to Merge
school_path_csv = Path("Resources/schools_complete.csv")
student_path_csv = Path("Resources/students_complete.csv")

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_path_csv)
student_data = pd.read_csv(student_path_csv)

# Combine the data into a single dataset.  
school_data_complete_df = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete_df.head()
```

```python
#Average Math Score
mean_reading_score = school_data_complete_df["reading_score"].mean()
mean_reading_score
```

```python
#Average Reading Score
mean_math_score = school_data_complete_df["math_score"].mean()
mean_math_score
```

```python
#Average Overall Score
mean_overall_score = (mean_reading_score + mean_math_score)/2
mean_overall_score
```

```python
# Create bins for the student reading grades
reading_bins = [0, 59.5, 69.5, 79.5, 89.5, 100]

# Create labels for the student reading grades
reading_group_labels = ["F", "D", "C", "B", "A"]
```

```python
# Place the student reading grades into bins
pd.cut(school_data_complete_df["reading_score"], reading_bins, labels=reading_group_labels)
```

```python
# Place the reading data series into a new column inside of the DataFrame
#school_data_complete_df.insert(column=6,["reading_score_grade"])
school_data_complete_df["reading_score_grade"] = pd.cut(school_data_complete_df["reading_score"], reading_bins, labels=reading_group_labels, include_lowest=True)
school_data_complete_df.head()
```

```python
# Create bins for the student math grades
math_bins = [0, 59.5, 69.5, 79.5, 89.5, 100]

# Create labels for the student math grades
math_group_labels = ["F", "D", "C", "B", "A"]
```

```python
# Place the student math grades into bins
pd.cut(school_data_complete_df["math_score"], math_bins, labels=math_group_labels)
```

```python
# Place the math data series into a new column inside of the DataFrame
#school_data_complete_df.insert(column=6,["reading_score_grade"])
school_data_complete_df["math_score_grade"] = pd.cut(school_data_complete_df["math_score"], math_bins, labels=math_group_labels, include_lowest=True)
school_data_complete_df.head()
```

```python
# Create a GroupBy object based upon "reading_score_grade"
reading_score_grade_group = school_data_complete_df.groupby("reading_score_grade")

# Find how many rows fall into each bin
print(reading_score_grade_group["reading_score"].count())
```

```python
# Create a GroupBy object based upon "math_score_grade"
math_score_grade_group = school_data_complete_df.groupby("math_score_grade")

# Find how many rows fall into each bin
print(math_score_grade_group["math_score"].count())
```

```python
#Calculate the reading and math passing with a score of 60 or more
school_data_complete_df["passing_reading"] =school_data_complete_df["reading_score"] >=60
school_data_complete_df["passing_math"] =school_data_complete_df["math_score"] >=60
passing_reading = ((school_data_complete_df["passing_reading"]).mean())*100
passing_reading
```

```python
passing_math = ((school_data_complete_df["passing_math"]).mean())*100
passing_math
```

```python
#Calculate the students that passed both reading and math
overall_passing = (passing_reading + passing_math)/2
overall_passing
```

```python
#Create the Summary Results DataFrame
school_results = [{"Total Schools": total_schools, 
            "Total Students": total_students, 
            "Total Budget": total_budget, 
            "Average Reading Score":  round(mean_reading_score,2), 
            "Average Math Score":  round(mean_math_score,2), 
           "Passing Reading": round(passing_reading,2),
           "Passing Math": round(passing_math,2),
            "Overall Passing": round(overall_passing,2)}]
#school_results.head()
school_results = pd.DataFrame(school_results)
school_results
```
